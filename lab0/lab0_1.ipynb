1. Kiểm tra xem chuỗi có chứa cả chữ cái và chữ số không.

```python
is_num = 0
is_char = 0

for i in range(len(str)):
    if str[i].isdigit():
        is_num += 1
    elif str[i].isalpha():
        is_char += 1
    if is_num and is_char: 
        print("Yes")
        break

if is_num==0 or is_char==0:
    print("No")
```

```python
import math
```

```python
n = 14
```

a. Kiểm tra xem n có phải số nguyên tố không.

```python
def is_prime(n):
    if n < 2: return False
    if n >= 4:
        for i in range(2, int(math.sqrt(n))):
            if n % i == 0: return False
        return True

print(is_prime(n))
```

b. In ra tất cả các ước số của n.

```python
a = []
for i in range(1, int(n/2) + 1):
    if n%i==0: a.append(i)

print(a)
```

c. Tìm tổng các chữ số của n.

```python
def sum_element(n):
    if n == 0: return 0
    s = 0
    while n:
        s += n%10
        n //= 10

    return s

sum_element(n)
```

```python
roman_numberal = {'I': 1, 'V': 5, 'X': 10, 'L': 50, 'C': 100, 'D': 500, 'M': 1000}

roman = 'XXXVI'
num = 0
for i in range(len(roman)):
    if i + 1 < len(roman) and  roman_numberal[roman[i]] < roman_numberal[roman[i + 1]]:
        num -= roman_numberal[roman[i]]
    else:   
        num += roman_numberal[roman[i]]
    
num
```

# Bai 7

```python
height = [1,8,6,2,5,4,8,3,7]

l = 0
r = len(height) - 1
max_a = 0
while l <= r:
    w = r - l
    h = min(height[l], height[r])
    area = w * h
    if  area > max_a:
        max_a = w*h
    if height[l] > height[r]:
        r -= 1
    else:
        l += 1

max_a
```

# Bai 8

```python
num = 19

def happy_number(n):
    if n <= 5: return False

    while n > 5:
        s = 0
        while n != 0:
            s += (n % 10) ** 2
            n //= 10
        if s == 1: return True
        else: 
            n = s
    return False

happy_number(num)
```

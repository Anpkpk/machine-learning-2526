# Bai 5

```python
arr = np.random.normal(loc=0, scale=1, size=(1000))
```

```python
e = np.mean(arr)
v = np.var(arr)
std_1 = np.std(arr) 
e, v, std_1
```

```python
arr = np.sort(arr)
arr = np.around(arr, 3)
```

```python
n = np.size(arr)
mid = n // 2
mid_left = mid // 2
mid_right = (n + mid) // 2

n, mid, mid_left, mid_right
```

```python
q1 = arr[mid_left]
q2 = arr[mid]
q3 = arr[mid_right]

q1, q2, q3
```

```python
iqr = q3 - q1
iqr
```

```python
import matplotlib.pyplot as plt
```

```python
plt.hist(arr)
plt.show()
```

```python
plt.boxplot(arr)
plt.show()
```

```python
l = q1 - 1.5*iqr
r = q3 + 1.5*iqr 
    
new_arr = arr[(arr > l) & (arr < r)]

e_2 = np.mean(new_arr)
v_2 = np.var(new_arr)
std_2 = np.std(new_arr)

e_2, v_2, std_2
```

```python
plt.hist(new_arr)
plt.show()
```

```python
plt.boxplot(new_arr)
plt.show()
```

```python
print(f'E1: {e}, E2: {e_2}')
print(f'V1: {v}, V2: {v_2}')
print(f'std1: {std_1}, std2: {std_2}')
```

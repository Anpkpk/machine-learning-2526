$f(x, y) = (x−2)^2 + (y + 1)^2 + 3, \quad (x_0, y_0) = (0, 0), \quad \alpha = 0.1$

```python
def f(x, y):
    return (x - 2)**2 + (y + 1)**2 + 3

def dfdx(x, y):
    return 2*(x - 2)

def dfdy(x, y):
    return 2*(y + 1)
```

```python
lr = 0.1

x, y = 0.0, 0.0

for i in range(5):
    gx = dfdx(x, y)
    gy = dfdy(x, y)
    
    x = x - lr * gx
    y = y - lr * gy
    
    f_val = f(x, y)
    grad_norm = gx**2 + gy**2  
    
    print(f"Step {i+1}: (x, y) = ({x:.2f}, {y:.2f}), "
          f"f = {f_val:.2f}, "
          f"gradient norm = {grad_norm:.2f}")
```

```python
def predict(X, w0, w1):
    y_pred = w0 + w1*X
    return y_pred
```

```python
def loss_function(y_true, y_pred):
    n = y_true.shape[0]
    s = np.sum((y_true - y_pred)**2)
    return (1/n) * s
```

```python
def gradient(X, y, w0, w1):
    y_pred = predict(X, w0, w1)
    n = y_pred.shape[0]

    dw0 = -2/n * np.sum(y - y_pred)
    dw1 = -2/n * np.sum((y - y_pred)*X)

    return dw0, dw1
```

```python
def fit(X, y, learning_rate=0.01, epochs=1000):
    losses = []
    w0, w1 = 0.0, 0.0

    for epoch in range(epochs):
        dw0, dw1 = gradient(X, y, w0, w1)

        w0 -= learning_rate*dw0
        w1 -= learning_rate*dw1

        y_pred = predict(X, w0, w1)
        loss = loss_function(y, y_pred)
        losses.append(loss)
        if epoch % 100 == 0:
            print(f'Epoch {epoch}, loss: {loss}')

    return w0, w1, losses
```

```python
def plot_learning_curve(losses):
    plt.figure(figsize=(10, 6))
    plt.plot(range(len(losses)), losses, label='Training Loss (MSE)')
    plt.xlabel('Epochs')
    plt.ylabel('MSE Loss')
    plt.title('Learning Curve')
    plt.legend()
    plt.grid()
    plt.show()
```

```python
w0, w1, losses = fit(X_train, y_train, learning_rate=0.05, epochs=1000)
plot_learning_curve(losses)
```

```python
def evaluate(X, y, w0, w1):
    y_pred = predict(X, w0, w1)
    
    mse = (1 / len(X)) * np.sum((y_pred - y)**2)
    rmse = np.sqrt(mse)

    mae = (1 / len(X)) * np.sum(abs(y - y_pred))
    
    y_mean = np.sum(y) / len(y)
    SS_res = np.sum((y - y_pred)**2)
    SS_tot = np.sum((y - y_mean)**2)
    
    r2 = (1 - SS_res/SS_tot)
    return mse, rmse, mae, r2
```

```python
class LinearRegression:
    def __init__(self, learning_rate=0.01, epochs=1000, bias=True):
        self.lr = learning_rate
        self.epochs = epochs
        self.weights = None
        self.bias = bias

    def predict(self, X):
        y_pred = np.dot(X, self.weights.T)
        return y_pred

    def loss_function(self, y_true, y_pred):
        S = np.sum((y_true - y_pred) ** 2)
        mse = S/len(y_true)
        return mse

    def gradient(self, X, y):
        y_pred = self.predict(X)
        dw = -(1/len(y))*np.dot(X.T, (y - y_pred))
        return dw
    
    def fit(self, X, y):
        self.weights = np.zeros(X.shape[1])
        losses = []
        for epoch in range(self.epochs):
            dw = self.gradient(X, y)
            self.weights -= self.lr * dw
            y_pred = self.predict(X)
            loss = self.loss_function(y, y_pred)
            losses.append(loss)
            if epoch % 100 == 0:
                print(f'Epoch {epoch}, Loss {loss}')

        return losses
```

```python
class LinearRegression:
    def __init__(self, learning_rate=0.01, epochs=1000, bias=True, batch_size=32, shuffle=True):
        self.lr = learning_rate
        self.epochs = epochs
        self.weights = None
        self.bias = bias
        self.shuffle = shuffle
        self.batch_size = batch_size

    def predict(self, X):
        y_pred = np.dot(X, self.weights.T)
        return y_pred

    def loss_function(self, y_true, y_pred):
        S = np.sum((y_true - y_pred) ** 2)
        mse = S/len(y_true)
        return mse

    def gradient(self, X, y):
        y_pred = self.predict(X)
        dw = -(1/len(y))*np.dot(X.T, (y - y_pred))
        return dw
    
    def batch_generator(self, X, y):
        if self.shuffle:
            idx = np.random.permutation(len(y))
            X, y = X[idx], y[idx]

        for i in range(0, len(y), self.batch_size):
            yield X[i:i+self.batch_size], y[i:i+self.batch_size]


    def fit(self, X, y):
        self.weights = np.zeros(X.shape[1])
        losses = []
        for epoch in range(self.epochs):
            for X_batch, y_batch in self.batch_generator(X, y):
                dw = self.gradient(X_batch, y_batch)
                self.weights -= self.lr * dw

            y_pred = self.predict(X)
            loss = self.loss_function(y, y_pred)
            losses.append(loss)
            if epoch % 100 == 0:
                print(f'Epoch {epoch}, Loss {loss}')

        return losses
```

```python
X = x
Y = y

X_scaled = (X - X.min())/(X.max() - X.min())
Y_scaled = (y - y.min())/(y.max() - y.min())
X_scaled = np.array(X_scaled)
Y_scaled = np.array(Y_scaled)

X_train, X_test, y_train, y_test = train_test_split(X_scaled, Y_scaled, test_size=0.2, random_state=42)

model = LinearRegression()
losses = model.fit(X_train, y_train)
plot_learning_curve(losses)
```
